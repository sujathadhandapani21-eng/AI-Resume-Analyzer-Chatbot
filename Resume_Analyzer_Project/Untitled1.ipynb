```python
import pandas as pd
import pdfplumber
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

resume_path = "sujatha.resume.pdf"

def extract_text(pdf_file):

    text = ""

    with pdfplumber.open(pdf_file) as pdf:

        for page in pdf.pages:

            extracted = page.extract_text()

            if extracted:
                text += extracted

    return text.lower()
resume_text = extract_text(resume_path)

print("RESUME TEXT")
print("--------------------------------")
print(resume_text)

skills_list = [
    "python",
    "sql",
    "machine learning",
    "data science",
    "excel",
    "power bi",
    "java",
    "html",
    "css",
    "c",
    "c++",
    "javascript"
]


found_skills = []

for skill in skills_list:

    if skill in resume_text:

        found_skills.append(skill)

print("\nSKILLS FOUND")
print("--------------------------------")
print(found_skills)


score = len(found_skills) * 10

print("\nRESUME SCORE")
print("--------------------------------")
print(score, "/ 100")


print("\nCHATBOT SUGGESTION")
print("--------------------------------")

if score >= 70:
    print("Excellent Resume")

elif score >= 40:
    print("Good Resume But Add More Skills")

else:
    print("Need Improvement")

job_description = """
Python SQL Machine Learning Data Science
"""


vectorizer = TfidfVectorizer()

vectors = vectorizer.fit_transform(
    [resume_text, job_description]
)

similarity = cosine_similarity(vectors)[0][1]

print("\nJOB MATCH PERCENTAGE")
print("--------------------------------")
print(round(similarity * 100, 2), "%")
```
